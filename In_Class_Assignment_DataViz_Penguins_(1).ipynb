## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
plt.figure(figsize=(8, 6))
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', s=80)
plt.title('Penguin Bill Dimensions by Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend(title='Species')
plt.show()
```

**Your interpretation (2–3 sentences):**

_Write your observations here._ The scatter plot shows clear separation between the three penguin species based on bill dimensions. Adelie penguins cluster in the lower right quadrant with shorter, deeper bills, while Gentoo penguins occupy the upper left with longer, shallower bills. Chinstrap penguins form a distinct group in the middle, showing that bill morphology is an effective characteristic for species identification.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins
plt.figure(figsize=(8, 5))
sns.histplot(data=penguins, x='bill_length_mm', bins=10, kde=True)
plt.title('Bill Length Distribution - 10 Bins')
plt.xlabel('Bill Length (mm)')
plt.show()
```

```python
# TODO: Histogram with 30 bins
plt.figure(figsize=(8, 5))
sns.histplot(data=penguins, x='bill_length_mm', bins=30, kde=True)
plt.title('Bill Length Distribution - 30 Bins')
plt.xlabel('Bill Length (mm)')
plt.show()
```

**Your comparison (1–2 sentences):**

_Write your observations here._ The 10-bin histogram shows a smoother, clearer bimodal distribution, while the 30-bin version reveals more detailed peaks and valleys but appears more fragmented. The coarser binning provides a better overall pattern, whereas finer binning shows more granular detail but with increased noise.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species
plt.figure(figsize=(8, 6))
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.title('Body Mass Distribution by Penguin Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (grams)')
plt.show()
```

**Your summary (1–2 sentences):**

_Write your observations here._ Gentoo penguins are clearly the heaviest species with a median body mass around 5000 grams, while Adelie and Chinstrap penguins are significantly lighter with median masses around 3700-3800 grams. The Gentoo penguins also show less variability in body mass and have a higher minimum weight compared to the other two species.
